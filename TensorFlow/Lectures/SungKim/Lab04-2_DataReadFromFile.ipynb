Loading data from file  
np.float32 자료형으로 불러옴  
numpy의 loadtxt 메소드 이용  
텍스트파일 전체를 읽어 ,로 데이터를 구분하고 읽어들임(csv는 ,로 구분된 벡터 데이터)  

Slicing:
xy에 csv파일 데이터테이블이 읽히고, slicing을 이용해서 x_data에는 xy의 모든 instance(row)에서 0번 열부터 마지막 열(-1) 전까지 읽고 저장  
y_data에는 모든 instance에서 마지막 열(-1)을 저장

```python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)  # for reproducibility

xy = np.loadtxt('data-01-test-score.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

데이터 확인

```python
# Make sure the shape and data are OK
print(x_data, "\nx_data shape:", x_data.shape)
print(y_data, "\ny_data shape:", y_data.shape)

# data output
'''
[[ 73.  80.  75.]
 [ 93.  88.  93.]
 ...
 [ 76.  83.  71.]
 [ 96.  93.  95.]] 
x_data shape: (25, 3)
[[152.]
 [185.]
 ...
 [149.]
 [192.]] 
y_data shape: (25, 1)
'''
```

나머지 절차는 모두 동일

```python


# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis
hypothesis = tf.matmul(X, W) + b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], 
                                   feed_dict={X: x_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost:", cost_val, "\nPrediction:\n", hy_val)

# train output
'''
0 Cost: 21027.0 
Prediction:
 [[22.048063 ]
 [21.619772 ]
 ...
 [31.36112  ]
 [24.986364 ]]
10 Cost: 95.976326 
Prediction:
 [[157.11063 ]
 [183.99283 ]
 ...
 [167.48862 ]
 [193.25117 ]]
 1990 Cost: 24.863274 
Prediction:
 [[154.4393  ]
 [185.5584  ]
 ...
 [158.27443 ]
 [192.79778 ]]
2000 Cost: 24.722485 
Prediction:
 [[154.42894 ]
 [185.5586  ]
 ...
 [158.24257 ]
 [192.79166 ]]
'''
```

# Queue Runner 활용

1) filename_queue에 tf.train.string_input_producer()로 여러 파일들 읽어들이기  
2) tf.TextLineReader() 클래스와 read()메소드를 이용하여 키와 값을 따로 읽어들일 수 있다.  
-> 텍스트파일을 읽어들일 때 일반적으로 사용할 수 있다.  
3) record_defaults 옵션값으로 data type을 결정해주고 decode_csv()메소드로 값들을 노드로 가져올 수 있다.

```python
  
# Lab 4 Multi-variable linear regression
# https://www.tensorflow.org/programmers_guide/reading_data

import tensorflow as tf
tf.set_random_seed(777)  # for reproducibility

filename_queue = tf.train.string_input_producer(
    ['data-01-test-score.csv'], shuffle=False, name='filename_queue')

reader = tf.TextLineReader()
key, value = reader.read(filename_queue)

# Default values, in case of empty columns. Also specifies the type of the
# decoded result.
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)
```

batch 데이터 생성
batch: 일괄적으로 처리되는 데이터 집단  
한 batch에서 training example을 10개로 지정(batch size)  
여기서 한 batch는 data-01-test-score.csv파일이므로, 여기서 샘플을 10개(10행) 만든다는 의미  
batch() 메소드를 이용해 트레이닝셋을 만든다.  

```python
# collect batches of csv in
train_x_batch, train_y_batch = \
    tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)
print(train_x_batch) #10개의 instance를 가진 파일이 만들어짐.
#자료형은 위에서 record_defaults를 이용해 정의됨. 소수점 -> float32
#이걸 랜덤으로 뽑는건지?
```

기타사항 동일  
start_queue_runner()를 이용해 Queue Runner 실행. Session run Loop를 돌기 전에 실행함.  
train_x_batch, train_y_batch도 노드이므로 실행시켜야 함.  
x_batch, y_batch에 feed_dict를 이용해 줄 수도 있다.

```python

# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis
hypothesis = tf.matmul(X, W) + b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)


```

학습 과정

```python
# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())

# Start populating the filename queue.
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X: x_batch, Y: y_batch})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)

coord.request_stop()
coord.join(threads)
```

csv파일을 확인해보면, 0행부터 10개 가져왔다는걸 알 수 있음.  
batch 순서를 섞고자 할 시 shuffle_batch 메소드 활용 가능  
큐 처리를 위한 스레드를 선언한걸로 보이지만 정확히 어떤 절차인지는 모르겠음  

값 주고 테스트
