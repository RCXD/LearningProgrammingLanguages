### **6-2 셀의 삽입 방법**

1. 상단의 "+코드", "+텍스트"를 눌러 삽입합니다.
2. 코드 셀 위에 삽입(Ctrl + M A)

4. 마크다운 셀로 변경 (Ctrl + M M)

### **6-4 실행과 출력**

* 실행 : 재생 버튼을 클릭하여, 실행 (해당 셀 실행)
* Ctrl + Enter

1. Ctrl + Enter : 해당 셀을 실행하고 커서는 해당 셀에 남음
2. Shift + Enter : 해당 셀을 실행하고 커서는 다음 셀로 넘어감
3. Alt + Enter : 해당 셀을 실행하고 커서는 다음에 셀을 삽입하고 넘어감
