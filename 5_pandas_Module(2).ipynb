```
#4. 피벗테이블(pivot_table)
# - 피벗테이블은 엑셀의 피벗테이블 기능과 동일
# - 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여, 데이터를 조회하여 펼쳐놓은 것을 의미
# - 왼쪽에 나타나는 인덱스를 행 인덱스, 상단에 나타나는 인덱스를 열 인덱스라고 부름

pd.pivot_table(df, index='소속사', columns='혈액형', values='키') #인덱스(행)를 소속사로 두고, columns를 혈액형으로, 그 다음 값(테이블)을 키값으로 정리해줌
```

```
#5. GroupBy(그룹으로 묶기)
#groupby는 데이터를 그룹으로 묶어 분석할 때 활용
#- 소속사별 키의 평균, 성 별 키의 평균 등, 특정 그룹별 통계 및 데이터의 성질을 확인하고자 할 때 활용
df.groupby('소속사')
```

groupby와 함께 쓰는 집합함수
* count(): 갯수
* sum(): 합계
* mean(): 평균
* var(): 분산
* std(): 표준편차
* min(): 최소값
* max(): 최대값

```
df
```

```
df.groupby('소속사').count() #해당하는 값이 몇개인지 세어줌. NaN을 제외하고 세어준다.
```

```
#산술통계는 자동으로 산술통계가 가능한 열만 출력됨
df.groupby('그룹').mean() #그룹의 평균 
```

```
#특정 열만 출력하고 싶을 때
df.groupby('혈액형')['키'].mean()
```

```
#6. Multi-index(복합 인덱스)
#행 인덱스를 복합적으로 구성하고 싶은 경우는 인덱스를 리스트로 만들어 준다.
df.groupby(['혈액형', '성별']).mean()
```

```
# 5. apply
#apply는 Series나 DataFrame에 좀 더 구체적인 로직을 적용하고싶을 때 사용
# - apply를 적용하기 위해서는 함수가 먼저 정의되어야 한다.
# - apply는 정의한 로직 함수를 인자로 넘겨줌.

df.loc[df['성별']=='남자', '성별']=1
df.loc[df['성별']=='여자', '성별']=0
df
```

```
df=pd.read_csv('http://bit.ly/ds-korean-idol')

#apply에 사용할 함수를 정의
def male_or_female(x):
  if x== '남자':
    return 1
  if x=='여자':
    return 0

df['성별_NEW'] = df['성별'].apply(male_or_female)
df.head()
```

```
# "cm당 브랜드평판지수"를 구해보기(브랜드평판지수/키)
#열을 하나 만들어서, 'cm당 브랜드평판지수'로 만들어주고 함수식을 넣어서 출력하면 됨.
def cm_to_brand(df):
  return df['브랜드평판지수']*1.0/df['키']

df['cm당 브랜드평판지수']=df.apply(cm_to_brand, axis=1)
df
```

```
# 5-2 lambda 함수의 적용
# lambda는 1줄로 작성하는 간단한 함수식임.
# return을 별도로 표시하지 않음
#괴상하지만 익숙해지면 편하다고.

f = lambda x: 1 if x=='남자' else 0
df['성별'].apply(f)
```

```
df['키/2'] = df['키'].apply(lambda x: x/2)
df
```
